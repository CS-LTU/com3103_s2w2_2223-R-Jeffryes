```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(-10, 10, 0.01)
ys = np.sin(x)

plt.plot(x, ys, '-')

plt.title("Sample Visualization")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(-10, 10, 0.01)
ys = (np.sin(x) * np.sin(x))

plt.plot(x, ys, '-')

plt.title("Sample Visualization")
plt.show()
```
